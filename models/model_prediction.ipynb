```python
import joblib
import pandas as pd
```

```python
df = pd.read_csv('test.csv')
ids = df['ID']            
X = df.drop(columns=['ID']).values

##SVM
model = joblib.load('models/HNSC/SVM_label.pkl')
feature_cols = list(model.feature_names_in_)
X = df[feature_cols] 
y_pred = model.predict(X)
result = pd.DataFrame({
    'ID': ids,
    'y_pred': y_pred
})
result.to_csv('prediction_SVM.csv', index=False)


##GradientBoosting
model = joblib.load('models/HNSC/GradientBoosting_label.pkl')
y_pred = model.predict(X)
result = pd.DataFrame({
    'ID': ids,
    'y_pred': y_pred
})
result.to_csv('prediction_GradientBoosting.csv', index=False)

##LightGBM
model = joblib.load('models/HNSC/LightGBM_label.pkl')
y_pred = model.predict(X)
result = pd.DataFrame({
    'ID': ids,
    'y_pred': y_pred
})
result.to_csv('prediction_LightGBM.csv', index=False)

##XGBoost
model = joblib.load('models/HNSC/XGBoost_label.pkl')
y_pred = model.predict(X)
result = pd.DataFrame({
    'ID': ids,
    'y_pred': y_pred
})
result.to_csv('prediction_XGBoost.csv', index=False)
```
